```
df = pd.read_csv('train.csv')
df.head(5)
```

```
df2 = pd.read_csv('test.csv')
df2.head(5)
```

```
columns = list(df.columns)
X = np.asarray(df[[i for i in columns if i != 'label']])
X = X/255.0
Y = np.asarray(df['label'])
XtestReal = np.asarray(df2)
XtestReal = XtestReal/255.0
```

```
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.15, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```
classifier = svm.SVC(gamma=0.001)
#fit to the trainin data
classifier.fit(X_train,y_train)
```

```
yhat = classifier.predict(X_test)
print(round(accuracy_score(y_test,yhat),4))
print (classification_report(y_test, yhat))
```

```
classifier2 = svm.SVC(gamma=0.001)
classifier2.fit(X, Y)
```

```
yhattest = classifier2.predict(XtestReal)
```

```
pred_df = pd.DataFrame({'Label':yhattest})
pred_df = pred_df.reset_index(drop = False)
pred_df.rename(columns={'index':'ImageId'}, inplace=True)
pred_df['ImageId'] = pred_df['ImageId']+1
pred_df
```
